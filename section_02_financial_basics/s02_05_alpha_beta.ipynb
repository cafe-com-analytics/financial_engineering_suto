# Alpha and Beta

In financial engineering, ***Alpha*** and ***Beta*** stands for the coeficients of a linear regression.

Finance mode: $ r_i = \alpha + \beta*r_{M} $

Statistical learning mode: $ r_i = \beta_{0} + \beta_{1}*r_{M} $

The higher $ \alpha $ and $ \beta $, the better. It means the y-axis stock is outperforming the market.

```python
df_close = pd.read_parquet(interim_path+'data_acquired_2021_02_16.pqt')

lst_symbols = ['AAPL', 'GOOG', 'IBM', 'SBUX', 'SPY']

df_close_ret = pd.DataFrame()

for symbol in lst_symbols:
    df_close_ret[symbol+'_return'] = df_close[symbol].pct_change(1)

df_close_ret.dropna(how='all', inplace=True)
df_close_ret.head()
```

```python
df_close_ret.iloc[100:150, [1,-1]].plot(figsize=(10,5));
```
